```python
# 1. Set the variable test1 to the string 'This is a test of the emergency text system,' and save test1 to a file named test.txt.
print(("*" * 15 + "1" + "*" * 15))
test1 = "This is a test of the emergency text system"
with open("test.txt", "w") as file:
    file.write(test1)
    
with open("test.txt", "r") as file:
    print(file.read())

# 2. Read the contents of the file test.txt into the variable test2. Is there a difference between test 1 and test 2?
print(("*" * 15 + "2" + "*" * 15))
with open("test.txt", "r") as file:
    test2 = file.read()

print(f"test1: {test1}, test2: {test2}")
if test1 == test2:
    print("test1 == test2")   

# 3. Create a CSV file called books.csv by using these lines:
# title,author,year
# The Weirdstone of Brisingamen,Alan Garner,1960
# Perdido Street Station,China Miéville,2000
# Thud!,Terry Pratchett,2005
# The Spellman Files,Lisa Lutz,2007
# Small Gods,Terry Pratchett,1992
print(("*" * 15 + "3" + "*" * 15))
text = '''title,author,year
The Weirdstone of Brisingamen,Alan Garner,1960
Perdido Street Station,China Miéville,2000
Thud!,Terry Pratchett,2005
The Spellman Files,Lisa Lutz,2007
Small Gods,Terry Pratchett,1992
'''
with open('books.csv', 'w') as outfile:
    outfile.write(text)
    
# 4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with these fields: title (text), author (text), and year (integer).
print(("*" * 15 + "4" + "*" * 15))
import sqlite3
db = sqlite3.connect('books.db')
curs = db.cursor()
curs.execute('''create table book (title text, author text, year int)''')
db.commit()

# 5. Read books.csv and insert its data into the book table.
print(("*" * 15 + "5" + "*" * 15))
with open('books.csv', 'r') as infile:
    for line in infile:
        title, author, year = line.strip().split(',')
        curs.execute('''insert into book values (?, ?, ?)''', (title, author, year))
db.commit()

# 6. Select and print the title column from the book table in alphabetical order.
print(("*" * 15 + "6" + "*" * 15))
data = curs.execute('''select title from book order by title''')
for row in data:
    print(row[0])

# 7. From the book table, select and print all columns in the order of publication.
print(("*" * 15 + "7" + "*" * 15))
data1 = curs.execute('''select * from book order by year''')
for row in data1:
    print(row[0], row[1], row[2])


# 8. Use the sqlalchemy module to connect to the sqlite3 database books.db that you just made in exercise 6.
print(("*" * 15 + "8" + "*" * 15))
import sqlalchemy
engine = sqlalchemy.create_engine('sqlite:///books.db')
sql = 'select title from book order by title asc'
rows = engine.execute(sql)
for row in rows:
    print(row[0])

# 9. Install the Redis server and the Python redis library (pip install redis) on your computer. Create a Redis hash called test with the fields count (1) and name ('Fester Bestertester'). Print all the fields for test.
print(("*" * 15 + "9" + "*" * 15))
import redis
conn = redis.Redis()
conn.hset('test', {'count': 1, 'name': 'Fester Bestertester'})

# 10. Increment the count field of test and print it.
print(("*" * 15 + "10" + "*" * 15))
conn.hincrby('test', 'count', 3)
conn.hget('test', 'count')
```
